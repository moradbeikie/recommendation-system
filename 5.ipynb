```python
# Some functions for preprocessing text

#def remove_hashtag(row):
#    row = str(row)
 #   text = re.sub(r"#\w+", "", row)
 #   return text

def cleaningText(text):
    text = str(text)
    text = re.sub(r'{}?""@[A-Za-z0-9]+', ' ', text) # remove mentions
    text = re.sub(r'#[A-Za-z0-9]+', ' ', text) # remove hashtag
    text = re.sub(r'RT[\s]', ' ', text) # remove RT
    text = re.sub(r"http\S+", ' ', text)# remove link
    text = re.sub(r':[\s]', ' ', text)
    text = re.sub(r'id[\s]', ' ', text)
    text = re.sub(r'ns[\s]', ' ', text)
    text = re.sub(r':[\s]', ' ', text)
    text = re.sub(r',[\s]', ' ', text)
    text = re.sub(r'[0-9]+', ' ', text) # remove numbers
    text = re.sub(r'[??]+', ' ', text)
    text = re.sub(r'[(     )]+', ' ', text)
    text = re.sub(r"['']+", ' ', text)
    text = re.sub(r"[:]+", ' ', text)
    text = re.sub(r"[[]+", ' ', text)
    text = re.sub(r"[]]+", ' ', text)
    text = re.sub(r"[,]+", ' ', text)
    text = re.sub(r"[/]+", ' ', text)
    text = re.sub(r"[.]+", ' ', text)
    text = re.sub(r"[—]+", ' ', text)
    text = re.sub(r"[-]+", ' ', text)
    text = re.sub(r"[;]+", ' ', text)
    text = re.sub(r"[&]+", ' ', text)
    text = re.sub(r"[·]+", ' ', text)
    text = re.sub(r"[$]+", ' ', text)
    text = re.sub(r"[–]+", ' ', text)
    text = re.sub(r"[-]+", ' ', text)
    text = re.sub(r"[`]+", ' ', text)
    text = re.sub(r"[-]+", ' ', text)
    text = re.sub(r"[!]+", ' ', text)
    text = re.sub(r"['']+", ' ', text)
    text = re.sub(r"[@]+", '', text)
    text = re.sub(r"[+]+", ' ', text)
    text = re.sub(r"[{|}]+", ' ', text)
    text = re.sub(r'Category[\s]', ' ', text)
    text = re.sub(r'births[\s]', ' ', text)
    text = re.sub(r'"of""the"[\s]', ' ', text)
    text = re.sub(r'"for"|"of"|"the"|"from"|"in"|"all"|"with"[\s]', ' ', text)
#    text = re.sub(r"'from'\S+", ' ', text)
 #   text = re.sub(r"['from']+", ' ', text)

    text = text.replace('\n', ' ') # replace new line into space
#    text = text.translate(str.maketrans('', '', string.punctuation)) # remove all punctuations
    text = text.strip(' ') # remove characters space from both left and right text
    return text

def casefoldingText(text):# Converting all the characters in a text into lower case
#    text = str(text)
    text = text.lower()
    return text

def tokenizingText(text):# Tokenizing or splitting a string, text into a list of tokens
#    text = str(text)
    text = word_tokenize(text)
    return text

def filteringText(text): # Remove stopwors in a text
#    text = str(text)
    listStopwords = set(stopwords.words("english"))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered
    return text

def stemmingText(text): # Reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
    return text

def toSentence(list_words): # Convert list of words into sentence
    sentence = ' '.join(word for word in list_words)
    return sentence

def dup(text):
    l = text.split()
    k = []
    for i in l:

    # If condition is used to store unique string
    # in another list 'k'
        if (text.count(i)>=1 and (i not in k)):
            k.append(i)
    return(' '.join(k))
```

```python
def bigrams(words, bi_min=15, tri_min=10):
    bigram = gensim.models.Phrases(words, min_count = bi_min)
    bigram_mod = gensim.models.phrases.Phraser(bigram)
    return bigram_mod
```

```python
def get_corpus(df):
    """
    Get Bigram Model, Corpus, id2word mapping
    """
    bigram = bigrams(dataset.message)
    bigram = [bigram[review] for review in dataset.message]
    id2word = gensim.corpora.Dictionary(bigram)
    id2word.filter_extremes(no_below=19, no_above=0.35)
    id2word.compactify()
    corpus = [id2word.doc2bow(text) for text in bigram]
    return corpus, id2word, bigram
```

```python
train_corpus, train_id2word, bigram_train = get_corpus(dataset)
```
